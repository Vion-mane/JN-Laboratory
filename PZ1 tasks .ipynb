```python
# 3. Создание папки и проверка
print("\n3. Создание папки test_folder:")
!mkdir test_folder
print("\nСодержимое после создания папки:")
!dir
```

```python
# 4. Работа с файлами (создание, перемещение, удаление)
print("\n4. Работа с файлами:")

# Создаем тестовый файл
with open('test_file.txt', 'w') as f:
    f.write('test content')

# Перемещаем файл в папку
!move test_file.txt test_folder

# Проверяем содержимое папки
print("\nСодержимое папки test_folder:")
!dir test_folder

# Удаляем файл
!del test_folder\\test_file.txt

# Проверяем, что файл удален
print("\nПосле удаления файла:")
!dir test_folder
```
