```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 400)

y1 = x        # y = x
y2 = x ** 2   # y = x^2
y3 = x ** 3    # y = x^3

plt.figure(figsize=(10, 7))

plt.plot(x, y1, 'b--',  label='$y = x$',        linewidth=2)
plt.plot(x, y2, 'g-.',  label='$y = x^2$',      linewidth=2)
plt.plot(x, y3, 'r-',   label='$y = x^3$',      linewidth=2)

plt.title('Графики функций: $y = x$, $y = x^2$, $y = x^3$', fontsize=14)
plt.xlabel('Ось X', fontsize=12)
plt.ylabel('Ось Y', fontsize=12)

plt.axis('equal')

plt.grid(True, linestyle=':', alpha=0.7)
plt.legend(fontsize=12)

plt.axhline(0, color='black', linewidth=0.5) 
plt.axvline(0, color='black', linewidth=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

fig.colorbar(surf, shrink=0.5, aspect=10, label='Значение z')

ax.set_title(r'3D-график функции $z = \sin(\sqrt{x^2 + y^2})$', fontsize=14)
ax.set_xlabel('Ось X', fontsize=12)
ax.set_ylabel('Ось Y', fontsize=12)
ax.set_zlabel('Ось Z', fontsize=12)

ax.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y_linear = x
y_parabola = x**2
y_sin = np.sin(x)
y_cos = np.cos(x)

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

axs[0, 0].plot(x, y_linear, color='blue', linewidth=2)
axs[0, 0].set_title(r'Линейная функция: $y = x$')
axs[0, 0].grid(True)

axs[0, 1].plot(x, y_parabola, color='red', linewidth=2)
axs[0, 1].set_title(r'Парабола: $y = x^2$')
axs[0, 1].grid(True)

axs[1, 0].plot(x, y_sin, color='green', linewidth=2)
axs[1, 0].set_title(r'Синус: $y = \sin(x)$')
axs[1, 0].grid(True)

axs[1, 1].plot(x, y_cos, color='purple', linewidth=2)
axs[1, 1].set_title(r'Косинус: $y = \cos(x)$')
axs[1, 1].grid(True)

plt.suptitle('Четыре основных графика математических функций', fontsize=16)
plt.tight_layout()
plt.show()
```

***Задание 2***
Средняя зарплата разработчиков в месяц (тыс. $):

* Уровень: ['Junior', 'Middle', 'Senior', 'Lead', 'CTO']
* Зарплата (тыс. $): [2, 4, 7, 10, 15]

Используйте плавный градиент цвета от низших к высшим уровням.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

levels = ['Junior', 'Middle', 'Senior', 'Lead', 'CTO']
salaries = [2, 4, 7, 10, 15]

colors = ["#1f77b4", "#2ca02c"]
cmap = LinearSegmentedColormap.from_list("salary_gradient", colors, N=len(levels))
bar_colors = [cmap(i/(len(levels)-1)) for i in range(len(levels))]

plt.figure(figsize=(10, 6), dpi=100)
bars = plt.bar(levels, salaries, color=bar_colors, edgecolor='black', linewidth=1)

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'${height}K',
             ha='center', va='bottom',
             fontsize=12, fontweight='bold')

plt.title('Средняя зарплата разработчиков (тыс. $/мес)', fontsize=14, pad=20)
plt.xlabel('Уровень позиции', fontsize=12)
plt.ylabel('Зарплата (тыс. $)', fontsize=12)
plt.ylim(0, max(salaries)*1.2)

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.gca().spines[['top', 'right']].set_visible(False)

plt.tight_layout()
plt.show()
```

***Задание 3***
* Построить график подинтегральной функции.
* Вычислить площадь под кривой на заданном отрезке как значение
определенного интеграла.
* Закрасить область под графиком, чтобы визуализировать интеграл
* Найдите площадь под функцией:
$$
f(x) = x^2 \cdot e^{-x}
$$
на интервале [0, 3].



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def f(x):
    return x**2 * np.exp(-x)

result, error = quad(f, 0, 3)

x = np.linspace(0, 3, 500)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label=r'$f(x) = x^2 e^{-x}$')

plt.fill_between(x, y, color='skyblue', alpha=0.5, 
                label=f'Площадь = {result:.3f}')

plt.title(r'График функции $f(x) = x^2 e^{-x}$ на [0, 3]', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

integral_text = r'$\int_0^3 x^2 e^{-x} dx \approx $' + f'{result:.3f}'
plt.text(1.5, 0.2, integral_text,
        fontsize=14, bbox=dict(facecolor='white', alpha=0.8))

plt.show()

print(f'Значение интеграла: {result:.5f}')
print(f'Погрешность вычисления: {error:.2e}')
```

уместно).Визуализируйте:
$$ f(x, y) = e^{-x^2 - y^2} \cos\left(5(x^2 + y^2)\right)$$
на x, y ∈ [−3, 3]
* Дополнительный требования
* Используйте plot_surface() для плотных поверхностей.
* Используйте plot_wireframe() для каркасных моделей.
* Настроите цветовую карту через cmap .
* Добавьте подписи осей и заголовки

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

Z = np.exp(-X**2 - Y**2) * np.cos(5*(X**2 + Y**2))

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10, label='Значение z')
ax1.set_title(r'Поверхность функции\n$f(x,y) = e^{-x^2-y^2} \cos(5(x^2+y^2))$')
ax1.set_xlabel('Ось X')
ax1.set_ylabel('Ось Y')
ax1.set_zlabel('Ось Z')

ax2 = fig.add_subplot(122, projection='3d')
wire = ax2.plot_wireframe(X, Y, Z, color='blue', linewidth=0.5)
ax2.set_title(r'Каркасная модель функции\n$f(x,y) = e^{-x^2-y^2} \cos(5(x^2+y^2))$')
ax2.set_xlabel('Ось X')
ax2.set_ylabel('Ось Y')
ax2.set_zlabel('Ось Z')

plt.tight_layout()
plt.show()
```
